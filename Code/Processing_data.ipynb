# Sentiment Analysis using LLM

## Sentiment Label and Sentiment Score

```python
import pandas as pd
from transformers import pipeline

# Read CSV file from GitHub
url = 'https://raw.githubusercontent.com/Xintong1122/Decentraland_LLM/main/Data/message_cleaned.csv'
df = pd.read_csv(url)

# Extract data from "Content" column
content_data = df['Content'].tolist()

# Define sentiment analysis models to use
models = [
    "cardiffnlp/twitter-roberta-base-sentiment-latest",
    "lxyuan/distilbert-base-multilingual-cased-sentiments-student",
    "cardiffnlp/twitter-xlm-roberta-base-sentiment"
]

# Perform sentiment analysis for each model and store results
for idx, model_name in enumerate(models, start=1):
    try:
        sentiment_pipeline = pipeline("sentiment-analysis", model=model_name)

        # Debugging info: Check the length of each input text
        for text in content_data:
            print(f"Processing text of length {len(text)}")

        results = sentiment_pipeline(content_data)

        # Extract labels and scores
        labels = [result['label'] for result in results]
        scores = [result['score'] for result in results]

        # Add results to the dataframe
        df[f'LLM{idx}_Label'] = labels
        df[f'LLM{idx}_Score'] = scores
    except Exception as e:
        print(f"An error occurred with model {model_name}: {e}")

# Save results to a new CSV file
df.to_csv('message_with_sentiment.csv', index=False)
```

## Daily Average Sentiment

```python
import pandas as pd
import numpy as np

# Read data from file
file_path = 'https://raw.githubusercontent.com/Xintong1122/Decentraland_LLM/main/Data/message_with_sentiment.csv'  # Replace with your CSV file path
df = pd.read_csv(file_path)

# Check column names
print(df.columns)

# Define function to calculate sentiment score
def calculate_sentiment_score(label, score):
    if label == 'positive':
        return score
    elif label == 'negative':
        return -score
    else:
        return 0

# Calculate sentiment scores for each LLM
df['LLM1_Sentiment_Score'] = df.apply(lambda row: calculate_sentiment_score(row['LLM1_Label'], row['LLM1_Score']), axis=1)
df['LLM2_Sentiment_Score'] = df.apply(lambda row: calculate_sentiment_score(row['LLM2_Label'], row['LLM2_Score']), axis=1)
df['LLM3_Sentiment_Score'] = df.apply(lambda row: calculate_sentiment_score(row['LLM3_Label'], row['LLM3_Score']), axis=1)

# Calculate date
df['Date'] = pd.to_datetime(df['Date']).dt.date

# Calculate daily average sentiment scores
daily_sentiment_LLM1 = df.groupby('Date')['LLM1_Sentiment_Score'].mean().reset_index()
daily_sentiment_LLM2 = df.groupby('Date')['LLM2_Sentiment_Score'].mean().reset_index()
daily_sentiment_LLM3 = df.groupby('Date')['LLM3_Sentiment_Score'].mean().reset_index()

# Define labeling
def label_sentiment(score):
    if score > 0.1:  # Define threshold for positive sentiment
        return 'positive'
    elif score < -0.1:  # Define threshold for negative sentiment
        return 'negative'
    else:
        return 'neutral'

# Assign labels to daily average sentiment scores
daily_sentiment_LLM1['LLM1_Label'] = daily_sentiment_LLM1['LLM1_Sentiment_Score'].apply(label_sentiment)
daily_sentiment_LLM2['LLM2_Label'] = daily_sentiment_LLM2['LLM2_Sentiment_Score'].apply(label_sentiment)
daily_sentiment_LLM3['LLM3_Label'] = daily_sentiment_LLM3['LLM3_Sentiment_Score'].apply(label_sentiment)

# Merge results
daily_sentiment = daily_sentiment_LLM1.merge(daily_sentiment_LLM2, on='Date', how='outer').merge(daily_sentiment_LLM3, on='Date', how='outer')

# Save results to a new CSV file
output_file_path = 'daily_sentiment_scores.csv'  # Replace with the path where you want to save it
daily_sentiment.to_csv(output_file_path, index=False)
```
